```python
import pandas as pd
import numpy as np
import plotly.express as px
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
# Survived column -> 1 - Survided, 0 - died

test_table = pd.read_csv('test.csv')
train_table = pd.read_csv('train.csv')
train_table = train_table.rename(columns={'Pclass': 'Class'})
train_table.head()
# Colunas exclu√≠das do modelo: PassengerId, Name, Ticket, Cabin, SibSp, Parch
```

```python
# Replace NaN values in Age column with the age mean
train_table['Age'] = train_table['Age'].fillna(train_table['Age'].median())
# Replace NaN values in Embarked with S
train_table['Embarked'] = train_table['Embarked'].fillna('S')
```

```python
for column in train_table[['Class', 'Sex', 'Embarked']]:
  unique_values = train_table[column].unique()
  display(column)
  for value in unique_values:
    table = train_table[train_table[column] == value]

    display(value, table['Survived'].value_counts(normalize = True).map('{:.1%}'.format))
```

```python
# Analizing wich caracteristic from some columns impacted more on the survival rate
for column in train_table[['Class', 'Sex', 'SibSp', 'Parch', 'Embarked']]:
  graph = px.histogram(train_table, x=column, color='Survived')
  #exibir o gr√°fico
  graph.show()

# Columns that might influence the survival rate: 'Pclass', 'Sex', 'Embarked'
```

```python
for column in train_table:
  if train_table[column].dtype == 'object':
    display(column)

encoder = LabelEncoder()

for column in train_table:
  if train_table[column].dtype == 'object':
    train_table[column] = encoder.fit_transform(train_table[column])
display(train_table)
```

```python
# Colunas exclu√≠das do modelo: PassengerId, Name, Ticket, Cabin, SibSp, Parch

# x data and y data
  # y data I want to predict -> score_credito
  # x data used to make the prevision -> other data
# If column is not needed for the prediction it is better to remove it

y = train_table['Survived']
columns = ['PassengerId', 'Name', 'Ticket', 'Cabin', 'SibSp', 'Parch', 'Survived', 'Fare']
x = train_table.drop(columns = columns)

# split data in train and test (default -> train 70%, test 30%)

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3)
```

```python
# Step 3: Creat AI model -> Credit Score: Good, Standard or Poor
# Decision tree
# KNN

decisiontree_model = RandomForestClassifier()
knn_model = KNeighborsClassifier()

decisiontree_model.fit(x_train, y_train)
knn_model.fit(x_train, y_train)
```

```python
# Step 4: Choose the best model

decisiontree_prediction = decisiontree_model.predict(x_test)

# knn needs test data to be in numpy format
knn_prediction = knn_model.predict(x_test.to_numpy())

print(accuracy_score(y_test, decisiontree_prediction))
print(accuracy_score(y_test, knn_prediction))
```

```python
test_table = test_table.rename(columns={'Pclass':'Class'})


# Replace NaN values in Age column with the age mean
test_table['Age'] = test_table['Age'].fillna(test_table['Age'].mean())

test_table.info()
```

```python
encoder = LabelEncoder()

for column in test_table:
  if test_table[column].dtype == 'object':
    test_table[column] = encoder.fit_transform(test_table[column])
display(test_table)
```

```python
columns = ['PassengerId', 'Name', 'Ticket', 'Cabin', 'SibSp', 'Parch', 'Fare']

prediction = [decisiontree_model.predict(test_table.drop(columns = columns))]

display(prediction)
print(type(prediction))
```

```python
survived = pd.DataFrame(prediction).transpose()

test_table['Survived'] = survived
test_table = test_table[['PassengerId', 'Survived']]
test_table.to_csv('Survival_Prediction.csv', index=False)
```
